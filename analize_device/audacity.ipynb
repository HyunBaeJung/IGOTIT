```python
import pandas as pd
import re
```

```python
file_list=[]
dataframe_list=[]
for i in range(35):
    a='scream_text/spectrum'+str(i+1)+'.txt'
    file_list.append(a)
    b=pd.read_csv(a, sep='\t')
    dataframe_list.append(b)
    dataframe_list[i]=dataframe_list[i].rename(columns = {'주파수 (Hz)':'주파수 (Hz)'+str(i+1),'레벨 (dB)':'레벨 (dB)'+str(i+1)})
for i in range(35):
    if i==0:
        scream_analyze_table=pd.concat([dataframe_list[0],dataframe_list[1]],axis=1)
    if (i!=0 and i!=34):
        scream_analyze_table=pd.concat([scream_analyze_table,dataframe_list[i+1]],axis=1)
        
        
file_list=[]
dataframe_list=[]
for i in range(31):
    a='talk_text/spectrum'+str(i+1)+'.txt'
    file_list.append(a)
    b=pd.read_csv(a, sep='\t')
    dataframe_list.append(b)
    dataframe_list[i]=dataframe_list[i].rename(columns = {'주파수 (Hz)':'주파수 (Hz)'+str(i+1),'레벨 (dB)':'레벨 (dB)'+str(i+1)})
for i in range(31):
    if i==0:
        talk_analyze_table=pd.concat([dataframe_list[0],dataframe_list[1]],axis=1)
    if (i!=0 and i!=30):
        talk_analyze_table=pd.concat([talk_analyze_table,dataframe_list[i+1]],axis=1)
```
